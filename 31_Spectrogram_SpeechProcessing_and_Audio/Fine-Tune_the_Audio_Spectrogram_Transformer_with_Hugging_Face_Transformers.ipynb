```python
# pip install transformers[torch] datasets[audio] audiomentations
from datasets import load_dataset

esc50 = load_dataset("ashraq/esc50", split="train")

esc50
```

```python
import random
import IPython.display as ipd
import numpy as np



# Selecionar um exemplo aleatório
example = esc50[random.randint(0, len(esc50) - 1)]

# Acessar os dados de áudio e a taxa de amostragem
audio_data = example

audio_data
```

```python
# Acessar os dados de áudio e a taxa de amostragem
audio_data = example['audio']['array']

audio_data
```

```python
sample_rate = example['audio']['sampling_rate']

# Reproduzir o áudio
ipd.Audio(audio_data, rate=sample_rate)
```

```python
# Mais conciso:
# Selecionar um exemplo aleatório
example = esc50[random.randint(0, len(esc50) - 1)]

# Acessar os dados de áudio e a taxa de amostragem
audio_data = example['audio']['array']
sample_rate = example['audio']['sampling_rate']

# Reproduzir o áudio
ipd.Audio(audio_data, rate=sample_rate)
```
